Other imports and Synthetic dataset loading for training

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
import joblib
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

#Synthetic dataset loading.....
df = pd.read_csv("up_grievances_synthetic.csv")
print("Dataset Loaded Successfully!")
print("Total Records:", len(df))
print(df.head())

categories = df["category"].unique()
print("\nCategories found:", categories)
```

Prophet Model Training

```python
#Train Prophet models
models = {}
metrics = []

for category in categories:
    print(f"\nTraining Prophet model for category: {category}")

    df_cat = df[df["category"] == category].groupby("date")["count"].sum().reset_index()
    df_cat = df_cat.rename(columns={"date": "ds", "count": "y"})

    train = df_cat[:-60]
    test = df_cat[-60:]

    model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=True,
        daily_seasonality=False,
        seasonality_mode="additive"
    )
    model.fit(train)

    future = model.make_future_dataframe(periods=60)
    forecast = model.predict(future)

    forecast_tail = forecast.tail(60)
    y_true = test["y"].values
    y_pred = forecast_tail["yhat"].values

    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    metrics.append({"category": category, "MAE": mae, "RMSE": rmse})

    print(f"Model trained for {category} | MAE: {mae:.2f}, RMSE: {rmse:.2f}")

    models[category] = model

    model.plot(forecast)
    plt.title(f"{category} - Prophet Forecast")
    plt.xlabel("Date")
    plt.ylabel("Complaint Count")
    plt.show()
```

Saving Models

```python
joblib.dump(models, "up_forecast.pkl")
print("Saved combined Prophet models as 'up_forecast.pkl'")
```

Saving Metrics....

```python
metrics_df = pd.DataFrame(metrics)
metrics_df.to_csv("up_model_metrics.csv", index=False)
print("Model evaluation metrics:")
print(metrics_df)
```

For Validation that our Model is working perfect or not

```python
import pandas as pd

#our current metrics
metrics_df = pd.DataFrame({
    'category': ['Infrastructure', 'Utilities', 'Sanitation', 'Medical'],
    'MAE': [19.65, 18.47, 14.78, 12.22],
    'RMSE': [24.27, 23.14, 18.65, 15.80]
})

#Calculate additional metrics for each category
for category in metrics_df['category']:
    df_cat = df[df["category"] == category].groupby("date")["count"].sum().reset_index()

    mean_val = df_cat['count'].mean()
    std_val = df_cat['count'].std()

    mae = metrics_df[metrics_df['category'] == category]['MAE'].values[0]
    rmse = metrics_df[metrics_df['category'] == category]['RMSE'].values[0]

    #Calculate percentage errors
    mae_percent = (mae / mean_val) * 100
    rmse_percent = (rmse / mean_val) * 100

    print(f"\n{'='*60}")
    print(f"{category}")
    print(f"{'='*60}")
    print(f"Mean value:        {mean_val:.2f}")
    print(f"Std deviation:     {std_val:.2f}")
    print(f"MAE:               {mae:.2f} ({mae_percent:.1f}% of mean)")
    print(f"RMSE:              {rmse:.2f} ({rmse_percent:.1f}% of mean)")
    print(f"RMSE/MAE ratio:    {rmse/mae:.2f}")

    #Interpretation
    print(f"\nInterpretation:")
    if mae_percent < 5:
        print(" Excellent! Error < 5% of mean")
    elif mae_percent < 10:
        print(" Very Good! Error < 10% of mean")
    elif mae_percent < 15:
        print(" Good! Error < 15% of mean")
    elif mae_percent < 25:
        print(" Acceptable! Error < 25% of mean")
    else:
        print(" Needs Improvement! Error > 25% of mean")
```
